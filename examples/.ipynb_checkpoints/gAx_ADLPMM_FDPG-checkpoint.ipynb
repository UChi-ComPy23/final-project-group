```python
def run_solver(solver, K):
    """Run solver for K iterations and return time used."""
    t0 = time.time()
    for _ in range(K):
        solver.step()
    t1 = time.time()
    return solver, t1 - t0
```

### Case 1: Total Variation Denoisingï¼ŒADLPMM vs mis-specified FISTA baseline

```python
np.random.seed(0)
n = 300
K = 800 

# synthetic signal
x_true = np.zeros(n)
x_true[50:100] = 1.0
x_true[150:200] = -0.8

b = x_true + 0.1 * np.random.randn(n)
lam = 0.2

problem_tv = TotalVariationProblem(b, lam)
x0 = np.zeros(n)

# ADLPMM 
z0 = problem_tv.A(x0)
u0 = np.zeros_like(z0)
adl = ADLPMM(problem_tv, x0, z0, u0, lam, rho=1.0)
fista_tv = FISTA(problem_tv, x0, alpha=1e-3)

adl, t_adl = run_solver(adl, K)
fista_tv, t_fista = run_solver(fista_tv, K)

F_adl = np.array(adl.history["obj"])
F_fi = np.array(fista_tv.objs)

# approximate F* as the best value seen among all runs
F_star = min(F_adl.min(), F_fi.min())

gap_adl = F_adl - F_star
gap_fi = F_fi - F_star

plt.figure(figsize=(9,6))
plt.plot(gap_adl, label="ADLPMM")
plt.plot(gap_fi, label="FISTA")
plt.xlabel("Iteration")
plt.ylabel(r"$F(x_k) - F^*$")
plt.title("TV Denoising â€” Objective Gap")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Plot reconstruction
plt.figure(figsize=(9,6))
plt.plot(x_true, label="true")
plt.plot(adl.x, label="ADLPMM")
plt.plot(fista_tv.x, label="FISTA")
plt.title("TV Reconstruction")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)
n = 300
K = 800

x_true = np.zeros(n)
x_true[50:100]  = 1
x_true[150:200] = -0.8

b = x_true + 0.1 * np.random.randn(n)
lam = 0.2

problem_tv = TotalVariationProblem(b, lam)
x0 = np.zeros(n)

#solver
z0 = problem_tv.A(x0)
u0 = np.zeros_like(z0)
adl = ADLPMM(problem_tv, x0, z0, u0, lam=0.5, rho=0.02)
mu_smooth = 5e-3
L_smooth  = 1 + (problem_tv.normA**2) / mu_smooth
sfista = SmoothedFISTA(problem_tv,x0.copy(),mu=mu_smooth,L=L_smooth,lambda_g=1,lambda_h=0,monotone=False)

#run solvers
adl,   t_adl  = run_solver(adl,   K)
sfista, t_sfi = run_solver(sfista, K)

F_adl = np.array(adl.history["obj"])
F_sfi = np.array(sfista.objs)
F_star = min(F_adl.min(), F_sfi.min())

gap_adl = F_adl - F_star + 1e-16
gap_sfi = F_sfi - F_star + 1e-16

#plot
plt.figure(figsize=(9, 6))
plt.plot(gap_adl, label="ADLPMM")
plt.plot(gap_sfi, label="SFISTA")
plt.yscale("log")
plt.xlabel("Iteration")
plt.ylabel(r"$F(x_k) - F^*$")
plt.title("TV Denoising â€” ADLPMM vs SmoothedFISTA")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

x_adl = adl.x
x_sfi = sfista.x

plt.figure(figsize=(9, 3))
plt.plot(x_true, label="true", linewidth=2)
plt.plot(x_adl,  label="ADLPMM")
plt.plot(x_sfi,  label="SFISTA")
plt.title("TV Reconstruction")
plt.xlabel("Index")
plt.ylabel("x[i]")
plt.legend()
plt.tight_layout()
plt.show()

print("ADLPMM time: ", t_adl)
print("SmoothedFISTA: ", t_sfi)
```

### Case 3: Group Composite Lassoï¼Œminimize f(x) + g(Ax)

```python
np.random.seed(0)
n = 100
m = 120 
K = 600 

G = np.random.randn(m, n)
Q, _ = np.linalg.qr(G) # Q orthonormal columns
A = Q # A full column rank

# Build synthetic grouped signal
group_size = 6
groups = [np.arange(i, min(i + group_size, m)) for i in range(0, m, group_size)]
x_true = np.random.randn(n)
active_groups = [0, 3, 10]
for g in active_groups:
    idx = np.arange(g * group_size, (g + 1) * group_size)
    idx = idx[idx < n]
    x_true[idx] = np.random.randn(len(idx))

b = x_true + 0.05 * np.random.randn(n)
lam = 0.5

#adlpmm
problem = GroupLassoCompositeProblem(A, b, groups, lam)
x0 = np.zeros(n)
z0 = problem.A(x0)
u0 = np.zeros_like(z0)
adl = ADLPMM(problem, x0, z0, u0, lam=1, rho=1)

#sfista
mu_smooth = 1e-1
L_smooth = 1 + (problem.normA ** 2) / mu_smooth  # safe upper bound
sfista = SmoothedFISTA(problem, x0.copy(), mu=mu_smooth, L=L_smooth, lambda_g=1, lambda_h=0, monotone=False)

#fdpg
y0 = np.zeros(m)  # initial dual variable
L_dual = (problem.normA ** 2) / problem.mu
fdpg = FDPG(problem, x0=y0, L=L_dual)

adl, t_adl = run_solver(adl,   K)
sfista, t_sfi = run_solver(sfista, K)
fdpg, t_fdpg = run_solver(fdpg,  K)

F_adl = np.array(adl.history["obj"])
F_sfi = np.array(sfista.objs)
F_fdpg = np.array(fdpg.objs)
F_star = min(F_adl.min(), F_sfi.min(), F_fdpg.min())

gaps = {"ADLPMM":F_adl  - F_star + 1e-16, "SFISTA":F_sfi  - F_star + 1e-16,"FDPG":F_fdpg - F_star + 1e-16,}

plt.figure(figsize=(9, 6))
for name, vals in gaps.items():
    plt.plot(vals, label=name)
plt.yscale("log")
plt.title("Case 2 â€” Composite Group Lasso: Objective Gap")
plt.xlabel("Iteration")
plt.ylabel(r"$F(x_k) - F^*$")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print("ADLPMM time:", t_adl)
print("SFISTA time:", t_sfi)
print("FDPG time:", t_fdpg)
```
